```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = "img.jpg"
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Threshold
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Find all contours with hierarchy
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Separate external and internal contours
external_contours = []
internal_contours = []

for i, cnt in enumerate(contours):
    if hierarchy[0][i][3] == -1:  # no parent → external
        external_contours.append(cnt)
    else:  # has a parent → internal
        internal_contours.append(cnt)

# Draw separately
external_img = image.copy()
internal_img = image.copy()

cv2.drawContours(external_img, external_contours, -1, (0, 255, 0), 2)  # Green
cv2.drawContours(internal_img, internal_contours, -1, (0, 0, 255), 2)  # Red

# Convert to RGB
external_img_rgb = cv2.cvtColor(external_img, cv2.COLOR_BGR2RGB)
internal_img_rgb = cv2.cvtColor(internal_img, cv2.COLOR_BGR2RGB)

# Show results
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(external_img_rgb)
plt.title("External Contours (Green)")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(internal_img_rgb)
plt.title("Internal Contours (Red)")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image_path = "img.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Threshold
_, thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Find contours with hierarchy
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Blank masks
external_mask = np.zeros_like(image)
internal_mask = np.zeros_like(image)

# Separate external and internal contours
for i, cnt in enumerate(contours):
    if hierarchy[0][i][3] == -1:   # external
        cv2.drawContours(external_mask, [cnt], -1, 255, -1)  # filled white
    else:                          # internal
        cv2.drawContours(internal_mask, [cnt], -1, 255, -1)  # filled white

# Show results
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(external_mask, cmap="gray")
plt.title("External Output (mask)")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(internal_mask, cmap="gray")
plt.title("Internal Output (mask)")
plt.axis("off")

plt.show()
```
